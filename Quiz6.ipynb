```python
import pandas as pd
df = pd.read_csv('http://www.stat.wisc.edu/~jgillett/451/data/mtcars.csv', index_col=0)
```

```python
plt.boxplot(df['mpg'])
plt.title('Boxplot of mpg')
plt.ylabel('mpg')
plt.show()
```

```python
df['cyl'].value_counts().sort_index().plot(kind='bar')
plt.title('Counts of Cars by Cylinder')
plt.xlabel('Number of Cylinders')
plt.ylabel('Count')
plt.show()
```

```python
count_8_cylinders = df[df['cyl'] == 8].shape[0]
print(count_8_cylinders)
```

```python
# Histogram of 'mpg' column using specified bin edges
bins = [10, 15, 20, 25, 30, 35]
plt.hist(df['mpg'], bins=bins, edgecolor='black')
plt.title('Histogram of mpg')
plt.xlabel('mpg')
plt.ylabel('Count')
plt.show()
```

```python
hist, _ = np.histogram(df['mpg'], bins=bins)
max_count = max(hist)
print(max_count)
```

```python
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()  # create new blank figure
fig.suptitle('gridspec: three plots using 1/4, 1/4, and 1/2 of figure')
gs = fig.add_gridspec(nrows=4, ncols=1)  # grid of plot axes
x = np.linspace(start=0, stop=4*np.pi, num=100)  # fake data
xlim = (0, 13)  # common x-axis limits

ax1 = fig.add_subplot(gs[0, 0])  # first plot in top 1/4
ax1.plot(x, -2 + x/3, label=r'$y = -2 + \frac{x}{3}$')
ax1.set(xlim=xlim, ylim=(-2, 2))
ax1.xaxis.set_ticks([])  # suppress axis ticks & numbers
ax1.legend()

ax2 = fig.add_subplot(gs[1, 0])  # second plot in second 1/4
ax2.plot(x, np.cos(x), label=r'$y = \cos(x)$')
ax2.set(xlim=xlim, ylim=(-2, 2))
ax2.xaxis.set_ticks([])  # suppress axis ticks & numbers
ax2.legend()

ax3 = fig.add_subplot(gs[2:4, 0])  # third plot in bottom 1/2
ax3.plot(x, (-2 + x/3) + np.cos(x),
         label=r'$y=\left(-2 + \frac{x}{3}\right) + \cos(x)$')
ax3.set(xlim=xlim, ylim=(-2, 6))
_ = ax3.legend()

plt.show()
```

```python
import seaborn as sns
sns.boxplot(x="gear", y="mpg", data=df)
plt.title("Boxplots of mpg grouped by gear")
plt.show()
```

```python
for gear_value in [3, 4, 5]:
    sns.kdeplot(df[df['gear'] == gear_value]['mpg'], label=f"Gear {gear_value}")
plt.title("Density plots of mpg grouped by gear")
plt.legend()
plt.show()
```

```python
sns.scatterplot(x="wt", y="mpg", hue="gear", data=df, palette="viridis")
plt.title("Scatter plot of mpg vs. wt colored by gear")
plt.show()
```
